```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data1 = pd.DataFrame({
    'Traffic_Signal': ['Red', 'Green', 'Yellow', 'Red', 'Green', 'Yellow', 'Red', 'Green'],
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C']
})

cross_tab = pd.crosstab(data1['Category'], data1['Traffic_Signal'], normalize='index') * 100


ax = cross_tab.plot(kind='bar', stacked=True, colormap='viridis', figsize=(8, 5))

for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%', label_type='center', color='white', fontsize=10)

# Title and labels
plt.title('Traffic Signal Distribution (%)')
plt.xlabel('Category')
plt.ylabel('Percentage')
plt.legend(title='Traffic Signal')
plt.show()
```
